```
# 2) Loop and download
for slug in series:
    for name, geom in locations.items():
        params = {
            "g": geom,
            "stat": "mean",
            "freq": "D",         # annual; change to "M" for monthly, etc.
            "format": "csv"
        }
        url = f"https://api.cal-adapt.org/api/series/{slug}/events/"
        r = requests.get(url, params=params, timeout=60)
        r.raise_for_status()

        fname = out_dir / f"{slug}_{name}.csv"
        fname.write_text(r.text)
        print(f"Saved {fname}")
```
